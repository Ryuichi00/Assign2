```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits import mplot3d
%matplotlib inline
```

```python
df = pd.read_csv("fishdata2.csv", encoding='unicode_escape', header = None, low_memory = False)
df
```

```python
df = df.iloc[:, 0:12]
df
```

```python
header = ['year', 'month', 'day', 'hour', 'minute', 'second', 'carbon monoxide', 'PM 2.5', 'external temperature', 'humidity', 'internal temperature', 'ignore']
df.columns = header
df
```

```python
df = df.dropna()
df
```

```python
df.dtypes
```

```python
df['year'] = df['year'].astype('int64')
df['month'] = df['month'].astype('int64')
df[['carbon monoxide' ]] = df[['carbon monoxide']].apply(pd.to_numeric, errors = 'coerce') 
df.dtypes
```

```python
df.describe()
```

```python
df = df[df['year'] > 2000]
df = df[df['year'] <= 2021]
df = df[df['month'] >= 1]
df = df[df['month'] <= 12]
df = df[df['external temperature'] > 0]
df
```

```python
df.describe()
```

```python
ex = df[0:40000]
ex
```

```python
from sklearn.cluster import MeanShift
model = MeanShift(bandwidth=50)
model.fit(ex)
pred = model.fit_predict(ex)
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(12, 10))
plt.scatter(ex["external temperature"], ex["humidity"], c=pred, cmap='plasma', alpha=0.9)
plt.show()
```

```python
fig = plt.figure(figsize=(10,10))
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
 
x = df['internal temperature']
y = df['external temperature']
z = df['PM 2.5']
 
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
 
ax.scatter3D(x, y, z, c = z, cmap = 'winter')
ax.set_xlabel('Internal Temperature')
ax.set_ylabel('External Temperature')
ax.set_zlabel('PM 2.5')

plt.show()
```

```python
Z = df.pivot_table(index='internal temperature', columns='external temperature', values='PM 2.5').T.values
x_unique = np.sort(x.unique())
y_unique = np.sort(y.unique())
X, Y = np.meshgrid(x_unique, y_unique)

fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection='3d')
ax.contour(X, Y, Z, 50, cmap='autumn')
ax.set_xlabel('Internal temperature')
ax.set_ylabel('External temperature')
ax.set_zlabel('PM 2.5')
ax.set_title('Surface')
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z, color = 'green')
ax.set_xlabel('Internal temperature')
ax.set_ylabel('External temperature')
ax.set_zlabel('PM 2.5')
ax.set_title('wireframe');
```
